#criando a matriz de correlação . codigo: https://seaborn.pydata.org/examples/many_pairwise_correlations.html

#A correlação, não quer dizer que é a causa 

#uma correlação alta é um caminho para explorar mais esta correlação

```python
import numpy as np
corr = dados.loc[:, 'g-0':'g-50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

#correlação de sobrevivencia celular. Todas estão bem correlacionadas. 

```python
corr_celular = dados.loc[:, 'c-0':'c-50'].corr()
```

```python
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_celular, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_celular, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

#criando uma coluna no final da tabela

```python
dados_resultados['n_moa'] = dados_resultados.drop('id', axis=1).sum(axis=1)
```

#Criando uma coluna no final da tabela para marcar aqueles compostos que tiveram alguma ação 

```python
dados_resultados['ativo_moa'] = dados_resultados['n_moa'] != 0
dados_resultados.head()
```

#Realizar o merge entre as duas tabelas
https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html

```python
dados_combinados = pd.merge(dados, dados_resultados[['id','n_moa', 'ativo_moa']], on = 'id')
dados_combinados.head()
```

```python
dados_combinados.query('tratamento == "com_droga"')['ativo_moa'].value_counts()
```

```python
#pegar os cinco compostos
composto_principal = dados_combinados['composto'].value_counts().index[:5]
plt.figure(figsize = [12, 8] )
#plotar os dados destes cinco compostos
sns.boxplot(data = dados_combinados.query('composto in @composto_principal'),y='g-0', x='composto', hue='ativo_moa')
```
