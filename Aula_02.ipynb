Renomear as coluna (drogas -> composto). O parametro inplace é necessário para alterar o nome no mapa

```python
mapa = {'droga':'composto'}
dados.rename(columns=mapa, inplace=True)
```

#pegar os 5 primeiros compostos

```python
cod_compostos = dados['composto'].value_counts().index[0:5]
```

```python
cod_compostos
```

#retornar os registros referente apenas aos 5 primeiros compostos


```python
dados.query('composto in @cod_compostos')
```

#plotar dados no gráfico seaborn  - frequencia do composto
#definir o tamanho do gráfico
#colocar um titulo
#show - para eliminar os comentarios em cima do gráfico

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize = [8, 6])
grafico = sns.countplot(x = 'composto', data = dados.query('composto in @cod_compostos'))
grafico.set_title('Top 5 Compostos -  frequência')
plt.show()
```
